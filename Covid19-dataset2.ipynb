```python
fig, ax = plt.subplots(figsize=(8,5))

# Add x-axis and y-axis
ax.scatter(df['Recovered'],
        df['Deaths'],
        color='purple'
       )

# Set title and labels for axes
ax.set(xlabel="Recovered",
       ylabel="Deaths",
       title="deaths vs recovered"
      )

plt.show()
```
